```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file = pd.read_csv('Sales.csv')

# Sample DataFrame (replace this with your actual data)
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Category_A': [200, 150, 300, 100],
    'Category_B': [300, 200, 250, 150],
    'Category_C': [400, 250, 350, 200]
}

df = pd.DataFrame(data)

# Set the index to Region
df.set_index('Region', inplace=True)

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt='g')

# Add title and labels
plt.title('Sales Performance Heatmap by Region and Product Category')
plt.xlabel('Product Category')
plt.ylabel('Region')

# Show the heatmap
plt.show()
```
